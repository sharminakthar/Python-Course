```python
#Styling dashboards

from matplotlib import pyplot as plt 
import seaborn as sns
%matplotliv inline
import warnings
warnings.filterwarnings('ignore')
```

```python
#Stylistic
#white, whitegrid, dark, darkgrid, ticks
sns.set_style=('dark')
sns.set(rc={'axes.facecolor' : 'black'})


f, axes = plt.subplots(2,2, figsize=(16,16))

#Plot 0,0
k1 = sns.kdeplot(x= movies.BudgetMillions, y= movies.AudienceRating, fill= True, palette='inferno', cmap='inferno', ax=axes[0,0],thresh=0)
#outline
k1b = sns.kdeplot(x= movies.BudgetMillions, y= movies.AudienceRating, fill= False,palette='cool', cmap='cool', ax=axes[0,0],thresh=0)


#k2 = sns.kdeplot(x= movies.BudgetMillions, y= movies.CriticRating, fill=True, cmap='mako', ax=axes[0,1])

#Plot 0,1
k3 = sns.kdeplot(x= movies.BudgetMillions, y =movies.CriticRating, fill= True, palette='inferno', cmap='inferno', ax=axes[0,1], thresh=0)
#outline
k3b = sns.kdeplot(x= movies.BudgetMillions, y =movies.CriticRating, fill= False,palette='cool', cmap='cool', ax=axes[0,1], thresh=0)

#Plot 1,0
v= sns.boxplot(data=movies[movies.Genre=='Drama'], x='Year', y='CriticRating', ax=axes[1,0], palette='cool')

#Plot 1,1
z= sns.violinplot(data=movies, x='Genre', y='CriticRating', ax=axes[1,1], palette='inferno')

k1.set(xlim=(-50,200))
k3.set(xlim=(-50,200))
k1b.set(xlim=(-50,200))
k3b.set(xlim=(-50,200))
```

```python
#Thematic Edits

list1= list()
mylabels = list()

sns.set_style=('whitegrid')
sns.axes_style("whitegrid")
sns.set(rc={'axes.facecolor' : 'white'})
for gen in movies.Genre.cat.categories:
    #assign / define
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)

fig, ax =plt.subplots()
h = plt.hist(list1, bins = 30, stacked = True, label= mylabels,
             rwidth=1)
plt.title('Movie Budget Distribution', fontsize= 35, color='DarkBlue',
          fontname='Console')
plt.ylabel('Number of Movies', fontsize=25, color='Red')
plt.xlabel('Budget', fontsize=25, color='Green')
plt.legend() 
plt.show()
```
